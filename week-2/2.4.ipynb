# AQ2.4

````python
shop_data = {
    'shop_num': [10, 10, 10, 11, 12, 11, 13, 12, 12],
    'item_num': [1002, 1003, 1004, 1001, 1001, 1002, 1003, 1004, 1003],
    'quantity': [100, 50, 80, 110, 100, 80, 100, 50, 40]
}

df1 = pd.DataFrame(shop_data)

item_data = {
    'item_num': [1001, 1002, 1003, 1004],
    'item_name': ['pen', 'notebook', 'pencil', 'eraser'],
    'price': [10, 30, 7, 15]
}

df2 = pd.DataFrame(item_data)
````

````python
# Display the original DataFrames without the index
print("Table Shop:")
display(HTML(df1.to_html(index=False)))

print("Table Item:")
display(HTML(df2.to_html(index=False)))
````

## Question 1

What will be the value of the ’Count’ returned by the following query?
```sql
SELECT COUNT(*)
FROM Shop, Item;
```

````python
# Shop, Item == cartesian product
cp_df = pd.merge(df1.assign(key=1), df2.assign(key=1), on='key').drop('key', axis=1)

cp_df
````

````python
cp_df.shape
````

Answer: Count * = 36

## Question 2

What will be the value of the ’Count’ returned by the following query?

```sql
SELECT COUNT(*)
FROM Shop, Item
WHERE Shop.item_num = Item.item_num;
```

````python
# select rows from cp_df where shop.item_num == item.item_num
result_df = cp_df[(cp_df.item_num_x == cp_df.item_num_y)]

print(result_df)

print(result_df.shape)
````

Answer: 9

## Question 3

What will be the value of the ’Count’ will be returned by the following query?

```sql
SELECT COUNT(*)
FROM Shop, Item
WHERE shop_num = 10;
```

````python
df3 = cp_df[(cp_df.shop_num == 10)]
print(df3)
print(df3.shape)
````

## Question 4

What will be the value of the ’Count’ will be returned by the following query?

```sql
SELECT COUNT(*)
FROM Shop, Item
WHERE shop_num = 10 and Shop.item_num = Item.item_num;
```

````python
df4 = cp_df[(cp_df.shop_num == 10) & (cp_df.item_num_x == cp_df.item_num_y)]
print(df4)
print(df4.shape)
````
